```python
import math

decode = {0:'a',1:'b',2:'c',3:'d',4:'e',5:'f',6:'g',7:'h',8:'i',9:'j',10:'k',11:'l',12:'m',13:'n',14:'o',
         15:'p',16:'q',17:'r',18:'s',19:'t',20:'u',21:'v',22:'w',23:'x',24:'y',25:'z',26:'A',27:'B',28:'C',29:'D',
          30:'E',31:'F',32:'G',33:'H',34:'I',35:'J',36:'K',37:'L',38:'M',39:'N',40:'O',41:'P',42:'Q',43:'R',44:'S',
          45:'T',46:'U',47:'V',48:'W',49:'X',50:'Y',51:'Z',52:'0',53:'1',54:'2',55:'3',56:'4',57:'5',58:'6',59:'7',
          60:'8',61:'9',62:' ',63:'.',64:'!',65:'?',66:',',67:':',68:'\'',69:'\"',70:'+',71:'-',72:'='}

encode = {v: k for k, v in decode.items()}


def ternary(n):
    if n == 0:
        return '0'
    nums = []
    while n:
        n, r = divmod(n, 3)
        nums.append(str(r))
    return ''.join(reversed(nums))


def encodeString():
    string = input("Enter the string to encode: ")
    characters = [i for i in string]
    encodedList = ''
    encChar = ''
    for char in characters:
        encChar = str((ternary(int(encode[char]))))
        while len(encChar)<4:
            encChar = '0' + encChar
        encodedList = encodedList+encChar
    print(encodedList)


def convertToDecimal(n):
    num = ''
    if (n != 0):
        decimalNumber = 0
        i = 0
        remainder = 0
        while (n != 0):
            remainder = n % 10
            n = n // 10

            decimalNumber += remainder * math.pow(3, i)
            i += 1
        num += str(int(decimalNumber))
    else:
        return '0'
    return num


def decodeString():
    string = input("Enter the string to decode: ")
    characters = [string[idx:idx + 4] for idx in range(0, len(string), 4)]
    decodedString = ''
    decChar = ''
    for char in characters:
        decChar = convertToDecimal(int(char))
        decodedString = decodedString+decode[int(decChar)]
    print(decodedString)


def main():
    while True:
        inp = input('Enter 1 for encoding a string, 2 for decoding a string, or 3 to exit: ')
        if inp == '1':
            encodeString()
        elif inp == '2':
            decodeString()
        elif inp == '3':
            break
        else:
            print("Invalid input. Please enter 1, 2, or 3")


main()
```
